**1.3) Operators**

1.3.1) Arithmetic Operator

```python
print('--INT--')
print('3 + 2 =',3+2)
print('3 - 2 =',3-2)
print('3 / 2 =',3/2)
print('3 // 2 =',3//2)
print('3 % 2 =',3%2)
print('3 * 2 =',3*2)
print('3 ** 2 =',3**2)

print()

print('--STR--')
print('A + A =','A'+'A')
print('Rakshali * 4 =','Rakshali' * 4)

print()

print('--BOOL--')
print('OR')
print('True + True =',True + True) # True = 1; False = 0
print('True + False =',True + False)
print('False + True =',False + True)
print('False + False =',False + False)

print('-----')

print('AND')
print('True * True =',True * True)
print('True * False =',True * False)
print('False * True =',False * True)
print('False * False =',False * False)
```

1.3.2 Comparison Operators

```python
print('2 > 3 =', 2>3)
print('2 < 3 =', 2<3)
print('2 >= 3 =', 2>=3)
print('2 <- 3 =', 2<=3)
print('2 == 3 =', 2==3)
print('2 != 3 =', 2!=3)

print("2 == '2' =",2 == '2')
print("1 == True =",1 == True)
print("100 == True*100 =",100 == True*100)
```

1.3.3) Logical Operators

```python
print("--AND--")
print('True and True =',True and True)
print('True and False =',True and False)
print('False and True =',False and True)
print('False and False =',False and False)

print()

print("--OR--")
print('True or True =',True or True)
print('True or False =',True or False)
print('False or True =',False or True)
print('False or False =',False or False)

print()

print("--NOT--")
print("not(True) =",not(True))
print("not(False) =",not(False))
```

```python
# Write a Python function that checks a number and prints "Positive", "Negative", or "Zero" based on its value.

num = int(input("Enter a number:"))
if num > 0:
  print("Positive")
elif num == 0:
  print("Zero")
else:
  print("Negative")
```

```python
# Write a Python function that checks a grade and prints "A" for scores 90 and above, "B" for scores 80-89, "C" for scores 70-79, "D" for scores 60-69, and "F" for scores below 60.

marks = int(input("Enter marks:"))
if marks < 0 or marks > 100:
  print("Not a valid score.")
if marks >= 90 and marks <= 100:
  print("A")
elif marks >= 80 and marks < 90:
  print("B")
elif marks >= 70 and marks < 80:
  print("C")
elif marks >= 60 and marks < 70:
  print("D")
elif marks < 60:
  print("F")
```

```python
# Write a Python function that checks if a year is a leap year. A year is a leap year if it is divisible by 4 but not divisible by 100, except when it is divisible by 400.

year = int(input("Enter a year:"))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
  print("Year:",year,"is a leap year.")
else:
  print("Year:",year,"is not a leap year.")
```

```python
# Write a Python function that checks if a number is within a specific range (10 to 20 inclusive) and prints "In range" if it is, otherwise print "Out of range".

num = int(input("Enter a number:"))
if num >= 10 and num <= 20:
  print(num,"is in range from 10 to 20")
else:
  print(num,"is not in range from 10 to 20")
```

```python
#Write a Python function that checks if a character is a vowel. If it is, print "Vowel", otherwise print "Consonant".

char = input("Enter a character:")

for i in "aeiou":
  if i == char:
    print("Vowel")
    break
else:
  print("Consontant")
```

```python
# Write a Python one-liner that checks if a number is greater than 100 and assigns the string "Large" or "Small" based on the result.

num = int(input("Enter a number:"))
if num > 100:
  print("Large")
else:
  print("Small")
```

```python
# Write a Python function that takes a list of numbers and prints "All positive" if all numbers in the list are positive, otherwise print "Contains negative".

lst = [2, 5, 6, -9, 8, 34, -576]

for i in range(len(lst)):
  if lst[i] < 0:
    print("Contains negative")
    break
else:
  print("All positive")
```

```python
#Probablity of getting sum as 6
c = 0
num = int(input("Enter a number:"))
for i in range(1,7):
  for j in range(1,7):
    if i+j == num:
      print(i,j)
      c += 1
print("Probablity of getting the sum as",num,":",c,"/",36, "=",round(c/36,2))
```

```python
# probablity of getting sum as a number
for n in range(1,13):
  c = 0
  for i in range(1,7):
    for j in range(1,7):
      if i+j == n:
        c += 1
  print("Probablity of getting",n,"as a sum of 2 numbers:",round(c/36,2))
```
