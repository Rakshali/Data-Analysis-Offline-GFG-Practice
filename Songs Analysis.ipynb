2.2) Cleaning duration column and converting into seconds

```python
durations = []
for i in df.duration:
  lst = i.split(":")
  if(len(lst) == 2):
    durations.append(int(lst[0]) * 60 + int(lst[1]))
  elif(len(lst) == 3):
    durations.append(int(lst[0]) * 3600 + int(lst[1]) * 60 + int(lst[2]))

#print(durations)

df.duration = durations
```

Number of songs in specific duration

```python
bins = [0, 180, 240, 300, float('inf')]
durations_grouped = pd.cut(df['duration'], bins=bins, labels=['< 3 min', '3-4 min', '4-5 min', '> 5 min'])
duration_counts = durations_grouped.value_counts().reset_index()
duration_counts.columns = ['duration','count']
duration_counts
```

```python
duration_counts.plot(kind='bar', title = 'Distribution of Song Durations',x = 'duration', y = 'count')
```

```python
short_threshold = 180  # Seconds
long_threshold = 300  # Seconds
categorized = pd.cut(df['duration'], bins=[0, short_threshold, long_threshold, float('inf')],
                               labels=['Short (< 3 min)', 'Medium (3-5 min)', 'Long (> 5 min)'])
duration_categorized = categorized.value_counts().reset_index()
duration_categorized.columns = ['Duration Category','Count']
duration_categorized
```

```python
duration_categorized.plot(kind='bar', title = 'Distribution of Song Lengths',x = 'Duration Category', y = 'Count')
```
