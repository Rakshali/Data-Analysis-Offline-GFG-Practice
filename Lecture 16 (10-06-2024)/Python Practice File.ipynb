**1.3) Operators**

1.3.1) Arithmetic Operator

```python
print('--INT--')
print('3 + 2 =',3+2)
print('3 - 2 =',3-2)
print('3 / 2 =',3/2)
print('3 // 2 =',3//2)
print('3 % 2 =',3%2)
print('3 * 2 =',3*2)
print('3 ** 2 =',3**2)

print()

print('--STR--')
print('A + A =','A'+'A')
print('Rakshali * 4 =','Rakshali' * 4)

print()

print('--BOOL--')
print('OR')
print('True + True =',True + True) # True = 1; False = 0
print('True + False =',True + False)
print('False + True =',False + True)
print('False + False =',False + False)

print('-----')

print('AND')
print('True * True =',True * True)
print('True * False =',True * False)
print('False * True =',False * True)
print('False * False =',False * False)
```

1.3.2 Comparison Operators

```python
print('2 > 3 =', 2>3)
print('2 < 3 =', 2<3)
print('2 >= 3 =', 2>=3)
print('2 <- 3 =', 2<=3)
print('2 == 3 =', 2==3)
print('2 != 3 =', 2!=3)

print("2 == '2' =",2 == '2')
print("1 == True =",1 == True)
print("100 == True*100 =",100 == True*100)
```

1.3.3) Logical Operators

```python
print("--AND--")
print('True and True =',True and True)
print('True and False =',True and False)
print('False and True =',False and True)
print('False and False =',False and False)

print()

print("--OR--")
print('True or True =',True or True)
print('True or False =',True or False)
print('False or True =',False or True)
print('False or False =',False or False)

print()

print("--NOT--")
print("not(True) =",not(True))
print("not(False) =",not(False))
```

**1.4) Typecasting**

1.4.1) int

```python
print("float to int (26.10):",int(26.10))
print(type(int(26.10)))

print()

print("string to int ('123123'):",int('123123'))
print(type(int('123123')))

print()

print("bool to int (True):",int(True))
print(type(int(True)))
```

1.4.2) float

```python
print("int to float (26):",float(26))
print("string to float ('26'):",float('26'))
print("bool to float (True):",float(True))
```

1.4.3) string

```python
print("int to string (26):",str(26))
print("float to string (26.10):",str(26.10))
print("float to bool (True):",str(True))
```

1.4.4) bool

```python
print("int to bool (0):",bool(0))
print("int to bool (1):",bool(1))
print("float to bool (0.00000):",bool(0.00000))
print("float to bool (0.000000001):",bool(0.000000001))
print("string to bool ('0'):",bool('0'))
print("string to bool ('1'):",bool('1'))
print("string to bool ('0.00000000'):",bool('0.00000000'))
print("string to bool (''):",bool(''))
```
